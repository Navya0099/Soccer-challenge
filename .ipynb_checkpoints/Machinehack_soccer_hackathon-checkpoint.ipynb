```python
test['season'] = test['season'].astype('str')
test['league_id'] = test['league_id'].astype('str')
test['Team 1'] = test['Team 1'].astype('str')
test['Team2'] = test['Team2'].astype('str')
```

```python
data = pd.concat([train,test],axis =0)

data.drop(['importance1', 'importance2', 'score1',
       'score2', 'xg1', 'xg2', 'nsxg1', 'nsxg2', 'adj_score1', 'adj_score2','league'],axis=1, inplace=True)
data.head()
```

```python
#  targetencoder on team1, team2, leagueid, season..

from category_encoders import TargetEncoder
from sklearn.preprocessing import RobustScaler

encoder = TargetEncoder()
encode = TargetEncoder()

data['Team 1'] = encoder.fit_transform(data['Team 1'],data['Outcome'])
data['Team2'] = encode.fit_transform(data['Team2'],data['Outcome'])

data = pd.get_dummies(data, drop_first= True)

scaler = RobustScaler()
data[['SPI1', 'SPI2','proj_score1', 'proj_score2']] = scaler.fit_transform(data[['SPI1', 'SPI2','proj_score1', 'proj_score2']])

data.head()
```

```python
train1 = data.iloc[:len(train),:]
test1 = data.iloc[len(train):,:]
test1
```

```python
# Using algorithm that can handle missing values
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold,cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,BaggingClassifier,GradientBoostingClassifier,RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

x = train1.drop('Outcome',axis=1)
y = train1['Outcome']


model = KNeighborsClassifier()
m2 = DecisionTreeClassifier()
m3 = AdaBoostClassifier()
m4 = BaggingClassifier()
m5 = GradientBoostingClassifier()
m6 = RandomForestClassifier()

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

r1 = cross_val_score(model,x,y,cv=cv,scoring='f1').mean()
r2 = cross_val_score(m2,x,y,cv=cv,scoring='f1').mean()
r3 = cross_val_score(m3,x,y,cv=cv,scoring='f1').mean()
r4 = cross_val_score(m4,x,y,cv=cv,scoring='f1').mean()
r5 = cross_val_score(m5,x,y,cv=cv,scoring='f1').mean()
r6 = cross_val_score(m6,x,y,cv=cv,scoring='f1').mean()

print(r1,r2,r3,r4,r5,r6)
```

```python
skf = StratifiedKFold(n_splits=3,random_state=None,shuffle=False)

for train_index,test_index in skf.split(x,y):
    print("Train:",train_index,'Test:',test_index)
    X_train,X_test = x.iloc[train_index], x.iloc[test_index]
    y_train,y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
m4.fit(X_train,y_train)
m4.predict_proba(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix,accuracy_score,log_loss,classification_report


bc = BaggingClassifier()
bc.fit(X_train,y_train)
y_pred_lr = bc.predict(X_test)

cm = confusion_matrix(y_test,y_pred_lr)
acc = accuracy_score(y_test,y_pred_lr)*100
print('log loss is', log_loss(y_test,bc.predict_proba(X_test)))
print(classification_report(y_test,y_pred_lr))
print('accuracy of model is {}'.format(acc))
ConfusionMatrixDisplay(cm).plot(include_values=True)
plt.grid()
plt.show()
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
y_pred_lr = gb.predict(X_test)

cm = confusion_matrix(y_test,y_pred_lr)
acc = accuracy_score(y_test,y_pred_lr)*100
print('log loss is', log_loss(y_test,gb.predict_proba(X_test)))
print(classification_report(y_test,y_pred_lr))
print('accuracy of model is {}'.format(acc))
ConfusionMatrixDisplay(cm).plot(include_values=True)
plt.grid()
plt.show()
```
